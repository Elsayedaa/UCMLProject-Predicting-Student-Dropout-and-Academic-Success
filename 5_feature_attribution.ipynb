```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from feature_attribution import feature_attribution_summary, get_feature_labels
```

```python
target_labels = ["Dropout", "Enrolled", "Graduated"]
```

```python
pipeline_files = [
    'pls_xgb_A.pkl',
    'pls_xgb_B.pkl',
    'pls_xgb_C.pkl'
]
pipelines = {}
for file in pipeline_files:
    label = os.path.splitext(file)[0][-1]
    pipelines[label] = joblib.load(file)
```

### Model A

```python
fig, ax = plt.subplots(1,3, figsize = (12,4))

d = feature_attribution_summary('A', pipelines)
for comp in range(3):
   

    ax[comp].spines['top'].set_visible(False)
    ax[comp].spines['right'].set_visible(False)
    ax[comp].tick_params(labelsize = 10)
    
    fsc = np.argsort(np.abs(d['sorted_xrot'][:,comp]))[::-1]
    ax[comp].bar(
        np.arange(10),
        d['sorted_xrot'][fsc,comp][:10],
        color = '#861F41',
    )
    ax[comp].plot(np.linspace(-1,10,10), [0]*10, '#75787b', linewidth = 1.5)
    ax[comp].set_xlim(-1,10)
    ax[comp].set_ylim(-0.5,1.1)
    ax[comp].set_xticks(np.arange(10))
    ax[comp].set_xticklabels(get_feature_labels(d['feature_names'])[fsc][:10], rotation = 45, fontsize = 10, ha = 'right')
    if comp>0:
        ax[comp].set_yticks([])
    pos = ax[comp].get_position()

    subax = fig.add_axes([pos.x0+0.15, 0.675, 0.055, 0.2])
    subax.bar(np.arange(3), d['sorted_yload'][:,comp], color = '#E5751F')
    subax.set_xticks(np.arange(3))
    subax.set_xticklabels(target_labels, fontsize = 8, rotation = 45, ha = 'right')
    subax.spines['top'].set_visible(False)
    subax.spines['right'].set_visible(False)
    subax.set_ylim(-0.3,0.3)
    subax.set_title("Coef")

    ax[comp].set_title(f"PLS Component {comp+1}\n", fontsize = 12, y = 1.075)

ax[0].set_ylabel("Loading scores", fontsize = 12)
fig.subplots_adjust(wspace=0.3)
plt.suptitle("Feature loadings: Model A", y = 1.15, fontsize = 12)
```
